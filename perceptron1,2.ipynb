```python
from sklearn.linear_model import Perceptron
x=[[1,0],[1,1],[2,0]]
y=["boy","girl","boy"]
ml=Perceptron()
ml=ml.fit(x,y)
r=ml.predict([[1,0],[2,0]])
print(r)
```

```python
from sklearn.neural_network import MLPClassifier
x=[[1,0],[1,1],[2,0]]
y=["boy","girl","boy"]
m=MLPClassifier(hidden_layer_sizes=(10),activation="relu")
m=m.fit(x,y)
r=m.predict([[1,0],[1,1]])
print(r)
```
